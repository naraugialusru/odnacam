```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

#Our sentences we like to encode
sentences = ['This framework generates embeddings for each input sentence',
    'Sentences are passed as a list of string.', 
    'The quick brown fox jumps over the lazy dog.']

#Sentences are encoded by calling model.encode()
embeddings = model.encode(sentences)

#Print the embeddings
for sentence, embedding in zip(sentences, embeddings):
    print("Sentence:", sentence)
    print("Embedding:", embedding)
    print("")
    
```

```python
from numpy.linalg import norm

fub_std= lambda a, b: 1 - np.inner(a, b)/(norm(a)*norm(b))
sent_dist = lambda a, b: fub_std(model.encode(a), model.encode(b))
sent_angle = lambda a, b: 180*np.arccos(np.abs(1 - sent_dist(a, b)))/np.pi
```

```python
import requests
import json
url = 'http://convai.io/2017/data/train_full.json'
r = requests.get(url)
conversation_list = json.loads(r.content)

counter = 0
clean_conversations = []
for c in conversation_list:
    print(counter)
    counter += 1
    convo = []
    for t in c['thread']:
        speaker = t['userId']
        text = t['text']
        sentences = tokenize.sent_tokenize(text)
        ginger = [parser.parse(sentence)['result'] for sentence in sentences]
        convo.append((speaker, ginger))
    clean_conversations.append(convo)
```

```python
#!{sys.executable} -m pip install torch torchvision
import torch.nn as nn
import torch
import random

```

```python
num_input_features = 10
hidden_size = 20
num_layers = 2
batch_size = 3
sequence_length = random.randint(1, 100)
directionality = 2


rnn = nn.LSTM(num_input_features, hidden_size, num_layers, bidirectional=(directionality == 2))
input = torch.randn(sequence_length, batch_size, num_input_features)
h0 = torch.randn(directionality * num_layers, batch_size, hidden_size)
c0 = torch.randn(directionality * num_layers, batch_size, hidden_size)
output, (hn, cn) = rnn(input, (h0, c0))
```

```python
input.shape, output.shape, hn.shape, cn.shape
```

```python
embed_dim = directionality * hidden_size
attn_input = output # torch.randn(sequence_length, batch_size, embed_dim)
num_heads = 1

multihead_attn = nn.MultiheadAttention(embed_dim, num_heads)

query = attn_input
key = attn_input
value = attn_input

attn_output, attn_output_weights = multihead_attn(query, key, value)
```
